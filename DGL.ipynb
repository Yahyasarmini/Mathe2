```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from math import sin, cos
#
t=sp.Symbol('t')
x=sp.Function('x')(t)
xp=sp.Derivative(x)
xpp=sp.Derivative(xp)
#sol=sp.Eq(xpp+3*xp+2*x,0)
ode=sp.Eq(xpp+3*xp+2*x,0)
sp.dsolve(ode)
```

```python
t=sp.Symbol('t')
x=sp.Function('x')(t)
xp=sp.Derivative(x)
xpp=sp.Derivative(xp)
ode=sp.Eq(xpp+3*xp+2*x,2*sp.sin(t))
sp.dsolve(ode)
```

```python
t=sp.Symbol('t')
x=sp.Function('x')(t)
xp=sp.Derivative(x)
xpp=sp.Derivative(xp)
ode=sp.Eq(xpp+3*xp+2*x,2*sp.sin(t))
sp.dsolve(ode)
initconds = {x.subs(t, 0): 0, xp.subs(t, 0): 1}
sol = sp.dsolve(ode,ics=initconds)
sol
```
