```python
# @title
import matplotlib.pyplot as plt

# Daten fr alle Benchmarks
packet_sizes = [128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144, 524288]
mmap_bandwidth = [1624.57, 3348.98, 7211.36, 17905.44, 16587.05, 31436.10, 48029.63, 57485.01, 63953.01, 21126.36, 20832.81, 21203.85, 21530.94]
process_vm_bandwidth = [82.77, 172.55, 341.45, 544.63, 1135.93, 2492.07, 3708.65, 6630.20, 9588.21, 10630.57, 10667.04, 14474.27, 16514.51]
pipe_bandwidth = [34.62, 133.46, None, 491.93, None, 1996.12, None, 2254.36, None, 1533.72, None, 711.06, None]

# Paketgren in lesbaren Strings konvertieren
packet_sizes_labels = [str(size) for size in packet_sizes]

# Plot fr mmap erstellen
plt.figure(figsize=(8, 5))
plt.plot(packet_sizes_labels, mmap_bandwidth, marker='o', linestyle='-', color='blue')
plt.title('Bandbreite fr mmap Benchmark')
plt.xlabel('Paketgre (Bytes)')
plt.ylabel('Bandbreite (MB/s)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot fr process_vm_readv erstellen
plt.figure(figsize=(8, 5))
plt.plot(packet_sizes_labels, process_vm_bandwidth, marker='o', linestyle='-', color='green')
plt.title('Bandbreite fr process_vm_readv Benchmark')
plt.xlabel('Paketgre (Bytes)')
plt.ylabel('Bandbreite (MB/s)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
