```python
import numpy as np
import matplotlib.pyplot as plt
h = 0.001 # Schrittweite
t0 = 0.0 # Beginn der Zeitspanne
t1 = 10.0 # Ende der Zeitspanne
B=1.0
L_0=0.5
x_0=0.6
x_p=0.0
```

```python

def f(t,z):
    zp = np.array([z[1], -100 * (z[1] * z[0]**2 - B * z[1] + z[0] - L_0)])
    return zp
```

```python
def Euler(t0,t1,h,z0):
    n = int((t1-t0)/h)
    t = np.linspace(t0,t1,n+1)
    z = np.zeros([len(z0),n+1])
    z[:,0] = z0
    for k in range(n):
        z[:,k+1] = z[:,k]+h*f(t[k],z[:,k])
    return t, z
```

```python
z0 = np.array([x_0,x_p])
t, z = Euler(t0,t1,h,z0)
fig, (ax1) = plt.subplots(1)
ax1.plot(t,z[0,:],'-b')
ax1.set(xlabel='$t$', ylabel='$x$') # z_0 fÃ¼r z0 mit 0 tiefgestellt
fig, ax = plt.subplots()
ax.plot(z[0,:],z[1,:],'-b')
ax.set(xlabel='$x$', ylabel='$áº‹$')
```
